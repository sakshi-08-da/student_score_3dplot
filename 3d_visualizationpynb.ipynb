```python
import pandas as pd
import plotly.graph_objects as go
```

```python
path = "/content/Students Performance.csv"
df = pd.read_csv(path)
df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
columns_to_scale = ['Math_Score', 'Reading_Score', 'Writing_Score']
scaler = MinMaxScaler()
df_scaled = df.copy()
df_scaled[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])
```

```python
fig = go.Figure(data=[go.Scatter3d(
    x=df_scaled['Math_Score'],
    y=df_scaled['Reading_Score'],
    z=df_scaled['Writing_Score'],
    text=df_scaled['Placement_Score'],
    mode='markers',
    marker=dict(
        size=6,
        color=df_scaled['Math_Score'],
        colorscale='Blues',
        opacity=0.8
    )
)])

fig.update_layout(
    title='3D Scaled Student Scores',
    scene=dict(
        xaxis_title='Math Score',
        yaxis_title='Reading Score',
        zaxis_title='Writing Score'
    )
)
fig.show()
```
